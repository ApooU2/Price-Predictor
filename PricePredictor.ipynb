```python
#prediction of price surge on a bullish day
pnb["Price Day After"]= pnb["Close"].shift(-1)
```

```python
pnb["PriceIncrease"]= (pnb["Price Day After"] > pnb["Close"]).astype(int)
```

```python
pnb = pnb.loc["1990-01-01":].copy()
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1) 

#Cannot use cross validation (it will use future data to predict which is non-realistic)

# Last 100 rows to test rest all used to train
train = pnb.iloc[:-100]
test = pnb.iloc[-100:]

predictors=["Close","Volume","Open","High","Low"]
model.fit(train[predictors], train["PriceIncrease"])
```

```python
from sklearn.metrics import precision_score

#checking the performance using precision score

predic= model.predict(test[predictors])
```

```python
import pandas as pd
predic = pd.Series(predic, index=test.index)
```

```python
precision_score(test["PriceIncrease"],predic)
#The prediction score will be 
#the model is fairly inaccurate (odds are only 27% in favour of us)
```

```python
def predictor(train, test, predictors, model):
    model.fit(train[predictors], train["PriceIncrease"])
    predic = model.predict(test[predictors])
    predic = pd.Series(predic, index=test.index, name = "Predictions")
    combined = pd.concat([test["PriceIncrease"],predic], axis=1)
    return combined
    
```

```python
#Around 250 days of market openings in a year so we predict the data of 11th year using 10 year data 
def backTest(data, model, predictors, start= 250, step = 25):
    all_predictions=[]

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predictor(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
    
```

```python
predictions = backTest(pnb, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["PriceIncrease"], predictions["Predictions"])
```
